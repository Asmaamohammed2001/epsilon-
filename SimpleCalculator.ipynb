```python
import tkinter as tk
import math

class SimpleCalculator:

    def __init__(self, master):
        self.master = master
        master.title("Simple Addition Calculator")

        # Create input fields
        self.label1 = tk.Label(master, text="Number 1:")
        self.label1.grid(row=0, column=0)

        self.num1_entry = tk.Entry(master)
        self.num1_entry.grid(row=0, column=1)

        self.label2 = tk.Label(master, text="Number 2:")
        self.label2.grid(row=1, column=0)
                   
        self.num2_entry = tk.Entry(master)
        self.num2_entry.grid(row=1, column=1)

        # Create a button to perform addition
        self.add_button = tk.Button(master, text="Add", command=self.add_numbers)
        self.add_button.place(x=10, y=80)
        self.subtract_button = tk.Button(master, text="subtract", command=self.subtract_numbers)
        self.subtract_button.place(x=105, y=80)
        self.mulyplacate_button = tk.Button(master, text="mulyplacate", command=self.mulyplacate_numbers)
        self.mulyplacate_button.place(x=10, y=110)
        self.devide_button = tk.Button(master, text="devide", command=self.devide_numbers)
        self.devide_button.place(x=100, y=110)
        self.powered_button = tk.Button(master, text="powered", command=self.powered_numbers)
        self.powered_button.place(x=10, y=140)
        self.GCD_button = tk.Button(master, text="GCD", command=self.GCD_numbers)
        self.GCD_button.place(x=100, y=140)
        self.lcm_button = tk.Button(master, text="lcm", command=self.lcm_numbers)
        self.lcm_button.place(x=10, y=170)
        self.sin_button = tk.Button(master, text="sin", command=self.sin_numbers)
        self.sin_button.place(x=100, y=170)
        self.cos_button = tk.Button(master, text="cos", command=self.cos_numbers)
        self.cos_button.place(x=10, y=200)
        self.tan_button = tk.Button(master, text="tan", command=self.tan_numbers)
        self.tan_button.place(x=100, y=200)
        self.sqrt_button = tk.Button(master, text="sqrt", command=self.sqrt_numbers)
        self.sqrt_button.place(x=10, y=230)
        self.factorial_button = tk.Button(master, text="factorial", command=self.factorial_numbers)
        self.factorial_button.place(x=100, y=230)
        self.log_button = tk.Button(master, text="log", command=self.log_numbers)
        self.log_button.place(x=10, y=260)
        self.log10_button = tk.Button(master, text="log10", command=self.log10_numbers)
        self.log10_button.place(x=100, y=260)
        # Create a label to display the result
        self.result_label = tk.Label(master, text="Result:")
        self.result_label.place(x=10,y=300)

        self.result_value = tk.Label(master, text="")
        self.result_value.place(x=50,y=300)

    def add_numbers(self):
        try:
            num1 = float(self.num1_entry.get())
            num2 = float(self.num2_entry.get())
            result = num1 + num2
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def subtract_numbers(self):
        try:
            num1 = float(self.num1_entry.get())
            num2 = float(self.num2_entry.get())
            result = num1 - num2
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def mulyplacate_numbers(self):
        try:
            num1 = float(self.num1_entry.get())
            num2 = float(self.num2_entry.get())
            result = num1 * num2
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def devide_numbers(self):
        try:
            num1 = float(self.num1_entry.get())
            num2 = float(self.num2_entry.get())
            result = num1 / num2
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def powered_numbers(self):
        try:
            num1 = float(self.num1_entry.get())
            num2 = float(self.num2_entry.get())
            result =  math.pow (num1, num2)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def GCD_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            num2 = int(self.num2_entry.get())
            result = math.gcd(num1,num2)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def lcm_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            num2 = int(self.num2_entry.get())
            result = math.lcm(num1,num2)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def lcm_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            num2 = int(self.num2_entry.get())
            result = math.lcm(num1,num2)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def sin_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            result = math.sin(num1)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def cos_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            result = math.cos(num1)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def tan_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            result = math.tan(num1)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def sqrt_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            result = math.sqrt(num1)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def factorial_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            result = math.factorial(num1)
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def log_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            result = math.floor (math.log(num1))
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")
    def log10_numbers(self):
        try:
            num1 = int(self.num1_entry.get())
            result = math.ceil(math.log10(num1))
            self.result_value.config(text=str(result))
        except ValueError:
            self.result_value.config(text="Error")

# Set up the main application window
root = tk.Tk()
root.geometry('180x350')
calculator = SimpleCalculator(root)
root.mainloop()
```
